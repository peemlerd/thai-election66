1. **เพราะอะไรพรรคเบอร์ 1-6 ถึงได้สส.บัญชีรายชื่อ 1 คนทุกพรรค**  การเปลี่ยนกฏิกาเลือกตั้งจากบัตรใบเดียวเป็นบัตรสองใบทำให้ทุกพรรคต้องได้รับคะแนนบัญชีรายชื่อเพิ่มขึ้นจาก 70,000 คะแนนเป็น 350,000 คะแนนต่อสส.บัญชีรายชื่อ 1 ที่นั่ง กูรูการเมืองหลายท่านวิเคราะห์ว่าจะระบบบัตรสองใบจะทำให้พรรคจิ๋วที่ได้สส.บัญชีรายชื่อเพียง 1 คนในการเลือกตั้งปี 2562 ทั้งหมด 12 พรรค (หรือที่บางคนเรียกว่า *สส.ปัดเศษ*) หายไป สอดคล้องกับผลการเลือกตั้งปี 2566 ที่เหลือพรรคที่ไม่ชนะการเลือกตั้งแบ่งเขตเลยแต่ได้สส.บัญชีรายชื่อ 1 คนเพียง 7 พรรค ประกอบด้วยพรรค (เลขที่บัญชีรายชื่อ)

พรรคเสรีรวมไทยได้คะแนนเสียงกว่า 8 แสนเสียงในการเลือกตั้งปี 62 จึงไม่ใช่เรื่องแปลกที่จะได้รับคะแนนเสียงรวมกันกว่า 6 แสนเสียงในการเลือกตั้งปี 66 คำถามที่น่าสนใจกว่าคือทำไมพรรคเบอร์ 2 และเบอร์ 6 จึงได้คะแนนเสียงเพิ่มจากเดิมหลายเท่าตัว ไม่สูญพันธุ์เหมือนพรรคจิ๋วอื่นๆ และเราจะอธิบายคะแนนบัญชีรายชื่อของพรรคเบอร์ 1-6 ได้อย่างไร 

**หมายเหตุ:** อาจารย์ลอย ชุนพงษ์ทองและคุณภูมิใจ ฉัตรไมตรีเคยวิเคราะห์ปริมาณการซื้อเสียงโดยสันนิษฐานว่าคะแนนบัญชีรายชื่อของพรรคเบอร์ 1-6 เป็นอานิสงส์ของระบบบัตรสองใบ ผมจะศประเมินข้อสันนิษฐานดังกล่าวโดยละเอียด

2. **การที่พรรค "2 ลุง" ได้แก่พรรคพลังประชารัฐและพรรครวมไทยสร้างชาติส่งสส.แข่งกันทุกเขต ส่งผลให้เสียที่นั่งทั้งหมดกี่ที่นั่ง** นักวิเคราะห์การเมืองหลายท่านประเมินว่าจุดแข็งของพรรคพลังประชารัฐคือสส.เขต ส่วนจุดแข็งของพรรครวมไทยสร้างชาติคือความนิยมของพลเอกประยุทธ์ จันทร์โอชา ประเด็นที่น่าสนใจคือหาก "พรรคสองลุง" ไม่แยกกันในการเลือกตั้งครั้งที่ผ่านมา จะได้สส.ประมาณกี่ที่นั่ง 
    
3. **พรรคขั้วรัฐบาลเดิมตัดเสียงกันเอง หรือโหวตเชิงยุทธศาสตร์มากน้อยเพียงใด ปัจจัยเหล่านี้ส่งผลอย่างไรต่อจำนวนสส.แบ่งเขตของพรรคฝ่ายค้านเดิม** ช่วงก่อนเลือกตั้ง นักวิเคราะห์ส่วนมากทำนายว่าก้าวไกลมีโอกาสชนะการเลือกตั้งสส.เขตในเขต 1 ของทุกจังหวัด (อำเภอเมือง), กรุงเทพมหานคร และปริมณฑล
   
4. **บัตรเสียมีปัญหาอย่างไร ส่งผลอย่างไรบ้าง** หลายจังหวัดมีิอัตราบัตรเสียที่ต่างกันลิบลับ เป็นไปได้หรือไม่ที่เกณฑ์การวัดบัตรเสียขาดมาตรฐาน จะมีผลการเลือกตั้งกี่เขตที่เปลี่ยนไปหากบัตรเสียกลายเป็นบัตรดี
   
5. **วิเคราะห์อนาคต: หากประเทศไทยจัดการเลือกตั้งอีกครั้งปี 2570 (รัฐบาลอยู่ครบ 4 ปีและไม่เกิดการรัฐประหารอีกครั้ง)** ในการเลือกตั้งที่ผ่านมา บางเขตเลือกตั้งคะแนนของผู้ชนะกับผูู้ที่ได้รับคะแนนอันดับสองต่างกันน้อยมาก ปัจจัยใดบ้างที่จะส่งผลต่อการเลือกตั้งในครั้งต่อๆไป

```python
# Subset the partylist result to obtain the number of partylist votes each party obtains
id = 747
partylist_no = id_no[id]
party_result_id = result_party[result_party["party_id"] == id].set_index("cons_id")

# Identify the vote and the party affiliation constituent candidates with the same number receive 
candi_same_id = [cons_id + "_%s" %(int(partylist_no)) for cons_id in party_result_id.index.tolist()]

temp = result_candi[result_candi["mp_app_id"].isin(candi_same_id)].set_index("cons_id").drop(columns=["mp_app_rank","percent","mp_app_id"])
temp = temp.rename({"party_id":"party_id_CandidateConst", "mp_app_vote":"CandidateConst_vote"}, axis=1)
```

```python
# Join the dataframe of party list vote and candidate constituion vote
party_result_id = party_result_id.join(temp)
```

```python
# Modify the party color map so that every non-major party (i.e., doesn't win any candidate constituency seat)
# receives the same color
cons_lst = [701,706,707,709,743,763]
prog_lst = [705,726,762,719,740,773] 
for id in result_candi["party_id"].unique():
    if id not in cons_lst and id not in prog_lst:
        party_color_map[id] = "#000000"
```

```python
# Mark a district is economically prosperous (e.g., Bangkok, District 1)
status_lst = []
major_prov = ["BKK", "RYG", "CBI", "NBI", "PTE"]

for cons_id in party_result_id.index.tolist():
    prov = cons_id.split("_")[0]
    cons_no = int(cons_id.split("_")[1])
    if prov in major_prov:
        status_lst.append(2)
    elif(cons_no == 1):
        status_lst.append(1)
    else:
        status_lst.append(0)

party_result_id["status"] = status_lst
```

```python
def plotPartylistCandi(party_result_id,
                       figsize_tuple = (10,10), 
                       marker_dict = {0:".", 1:"+", 2:"*"}, 
                       alpha_dict = {0:1, 1:1, 2:1}, 
                       label_dict = {0:"Other", 1:"District 1", 2:"Major province"},
                       size_dict = {0:100, 1:100, 2:100},
                       id_lst = [701,705,743,763,726,709],
                       candi_bound = (0,80000),
                       party_bound = (0,5000),
                       color_map = party_color_map):
    # A function to create

    # Define the bound
    candi_lb, candi_ub = candi_bound[0], candi_bound[1]
    party_lb, party_ub = party_bound[0], party_bound[1]
    # Subset the dataframe
    temp = party_result_id[party_result_id["party_id_CandidateConst"].isin(id_lst)]
    temp = temp[(temp["CandidateConst_vote"] < candi_ub) & (temp["CandidateConst_vote"] > candi_lb)]
    temp =  temp[(temp["party_list_vote"] < party_ub) & (temp["party_list_vote"] > party_lb)]
    subset_df = temp.copy()

    # Create the necessary list of symbols
    status_lst = party_result_id["status"].unique()
    marker_lst = list(map(lambda x: marker_dict[x], status_lst))
    alpha_lst = list(map(lambda x: alpha_dict[x], status_lst))
    label_lst = list(map(lambda x: label_dict[x], status_lst))
    size_lst = list(map(lambda x: size_dict[x], status_lst))

    ax = plt.subplots(figsize = figsize_tuple)

     # Create the scatterplot for each of the cand
    for i, level in enumerate(status_lst):
        temp = subset_df[subset_df["status"] == level]
        plt.scatter(temp["CandidateConst_vote"], 
                    temp["party_list_vote"], 
                    c =temp["party_id_CandidateConst"].map(color_map),
                    marker = marker_lst[i],
                    label = label_lst[i],
                    s = size_lst[i],
                    alpha = alpha_lst[i])
            
    plt.xlabel("Votes for constituency candidate Number %i" %(partylist_no))
    plt.ylabel("Party list vote")
    plt.title("Party list votes and candidate constituency votes with the same number (%s)" %(int(partylist_no)))
    plt.xticks(np.arange(0,70000,10000))
    plt.yticks(np.arange(0,5000,250))
    plt.legend()
    plt.savefig("partylist_no_%s" %(3))
plotPartylistCandi(party_result_id, figsize_tuple = (8,8), color_map = party_color_map, candi_bound = (0,20000), party_bound = (0,2000))
```

แกนนอนของ plot ข้างบนคือคะแนนของสส.เขตเบอร์ 3 ส่วนแกนตั้งคือคะแนนบัญชีรายชื่อที่พรรคเป็นธรรม (เบอร์ 3) ได้รับ ส่วนสัญลักษณ์ต่างๆมีดังนี้

- สีแดง = เพื่อไทย; ส้ม = ก้าวไกล; ฟ้าอ่อน = ประชาธิปัตย์; น้ำเงิน = พลังประชารัฐ; และกรมท่า = รวมไทยสร้างชาติ

```python
ax = plt.subplots(figsize = (10,10))
marker_lst = ["+",".","*"]
alpha_lst = [1,1,1]
type = ["District 1", "Other", "Major province"]
size_lst = [100, 100, 100]


for i, level in enumerate(party_result_id["status"].unique()):
    temp = party_result_id[party_result_id["status"] == level]
    temp = temp[temp["party_id_CandidateConst"].isin(prog_lst)]
    plt.scatter(temp["CandidateConst_vote"], 
                temp["party_list_vote"], 
                c =temp["party_id_CandidateConst"].map(party_color_map),
                marker = marker_lst[i],
                label = type[i],
                s = size_lst[i],
                alpha = alpha_lst[i])
        
plt.xlabel("Votes for constituency candidate Number %i" %(partylist_no))
plt.ylabel("Party list vote")
plt.title("Party list votes and candidate constituency votes with the same number (%s)" %(int(partylist_no)))
plt.xticks(np.arange(0,70000,10000))
plt.yticks(np.arange(0,5000,250))
plt.legend()
```
