```python
def createOverlapDf(partyID, 
                    partyID_partylistNo, 
                    result_candi, 
                    result_partylist, 
                    major_prov = ["BKK","RYG","CBI", "SPK", "PTE","SKN","AYA", "NPT","NBI"]):
    """
    @partyID: the ID of a political party
    @partyID_partylistNo: the dictionary with key = party ID; value = party list number
    @result_candi: A dataframe of candidate results in all 400 constituencies
    @result_partylist: A dataframe of the party list results in all 400 constituencies 
    @major_prov: A list of major provinces
    Return: A dataframe consisting of party-list vote a party receives as well as the vote a candidate with the same number receives
    """
    # Create a dataframe consisting of parties from the same
    partylist_no = partyID_partylistNo[partyID]
    df = result_partylist[result_partylist["party_id"] == partyID].set_index("cons_id")
    
    # Identify the vote and the party affiliation constituent candidates with the same number receive 
    candi_same_id = [cons_id + "_%s" %(int(partylist_no)) for cons_id in df.index.tolist()]
    temp = result_candi[result_candi["mp_app_id"].isin(candi_same_id)].set_index("cons_id").drop(columns=["mp_app_rank","percent","mp_app_id"])
    temp = temp.rename({"party_id":"party_id_CandidateConst", "mp_app_vote":"CandidateConst_vote"}, axis=1)
    df = df.join(temp)

    # Define the status of each constituency
    status_lst = []
    for cons_id in df.index.tolist():
        prov = cons_id.split("_")[0]
        cons_no = int(cons_id.split("_")[1])
        if prov in major_prov:
            status_lst.append(2)
        elif(cons_no == 1):
            status_lst.append(1)
        else:
            status_lst.append(0)

    df["status"] = status_lst
    return df
```

```python
# Modify the party color map so that every small party is assigned the same color
party_color_map = dict(zip(info_party9["id"], info_party9["color"]))

for id in result_candi["party_id"].unique():
    if id not in cons_lst and id not in prog_lst:
        party_color_map[id] = "#000000"
```

```python
# Mark a district is economically prosperous (e.g., Bangkok, District 1)
party_result_id = createOverlapDf(747, partyID_partylistNo, result_candi, result_partylist)
```

```python
def plotPartylistCandi(party_result_id,
                       displayLine = True,
                       figsize_tuple = (10,10), 
                       marker_dict = {0:".", 1:"+", 2:"*"}, 
                       alpha_dict = {0:1, 1:1, 2:1}, 
                       label_dict = {0:"Other", 1:"District 1", 2:"Major province"},
                       size_dict = {0:100, 1:100, 2:100},
                       id_lst = [701,705,743,763,726,709,769,707,740,762],
                       candi_bound = (0,80000),
                       party_bound = (0,5000),
                       color_map = party_color_map,
                       title = "",
                       filename = ""):
    """
    A dataframe of candidate and party list result.
    """
    # A function to visualize the number of votes per 

    # Define the bound
    candi_lb, candi_ub = candi_bound[0], candi_bound[1]
    party_lb, party_ub = party_bound[0], party_bound[1]
    # Subset the dataframe
    temp = party_result_id[party_result_id["party_id_CandidateConst"].isin(id_lst)]
    temp = temp[(temp["CandidateConst_vote"] <= candi_ub) & (temp["CandidateConst_vote"] >= candi_lb)]
    temp =  temp[(temp["party_list_vote"] <= party_ub) & (temp["party_list_vote"] >= party_lb)]
    subset_df = temp.copy()

    # Create the necessary list of symbols
    status_lst = party_result_id["status"].unique()
    marker_lst = list(map(lambda x: marker_dict[x], status_lst))
    alpha_lst = list(map(lambda x: alpha_dict[x], status_lst))
    label_lst = list(map(lambda x: label_dict[x], status_lst))
    size_lst = list(map(lambda x: size_dict[x], status_lst))

    ax = plt.subplots(figsize = figsize_tuple)

     # Create the scatterplot for each of the cand
    for i, level in enumerate(status_lst):
        temp = subset_df[subset_df["status"] == level]
        plt.scatter(temp["CandidateConst_vote"], 
                    temp["party_list_vote"], 
                    c =temp["party_id_CandidateConst"].map(color_map),
                    marker = marker_lst[i],
                    label = label_lst[i],
                    s = size_lst[i],
                    alpha = alpha_lst[i])
    if(displayLine):
        y = subset_df["party_list_vote"]
        x = subset_df["CandidateConst_vote"]
        model = sm.OLS(y, sm.add_constant(x))
        results = model.fit()
        b, m = results.params
        plt.axline(xy1=(0, b), slope=m, label=f'$P = {m:.3f}C {b:+.3f}$')

    # Sanity check
    print("N major = %d" %(temp[temp["status"] == 2].shape[0]))
    
    plt.xlabel("Votes for constituency candidate Number %i (C)" %(partylist_no))
    plt.ylabel("Party list vote (P)")
    if(title == ""):
        plt.title("Party list votes and candidate constituency votes with the same number (%s)" %(int(partylist_no)))
    else:
        plt.title(title)
    plt.xticks(np.arange(0,70000,10000))
    plt.yticks(np.arange(0,5000,250))
    plt.legend()
    if(filename != ""):
        plt.savefig(filename)
```

```python
partyID = 747
filename = "partylist_no_%s.png" %(partyID_partylistNo[partyID])
party_result_id = createOverlapDf(partyID, partyID_partylistNo, result_candi, result_partylist)
plotPartylistCandi(party_result_id, 
                   figsize_tuple = (8,8), 
                   color_map = party_color_map, 
                   candi_bound = (0,60000), 
                   party_bound = (0,5000),
                   filename = filename)
```

```python
def plotSameProvince(party_result_id, pos, figsize = (8,8), displayLine = False):
    """
    Function: Create a plot of party list vote and candidate constituency vote for constituencies within the same province
              for constituencies with the same vote
    
    @param party_result_id: A dataframe of party-list vote and candidate constituency vote for a given ID
    @param pos: Position
    """
    top_vote = party_result_id.sort_values("party_list_vote", ascending=False)[pos:pos+1]
    provID = getProv(top_vote.index.tolist()[0])
    compare_prov = getComparison(top_vote.index.tolist(), party_result_id.index.tolist(), method = "province")
    #print(party_result_id.loc[compare_prov].shape[0], provID)
    plotPartylistCandi(party_result_id.loc[compare_prov], 
                       figsize_tuple = figsize, 
                       color_map = party_color_map, 
                       candi_bound = (0,80000), 
                       party_bound = (0,10000),
                       title = "%s: Party-list vs candidate constituency with same number" %(provID_provName[provID]),
                       displayLine = displayLine)
```
